```python
# Defining the needed libraries
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python

# Generate synthetic data
np.random.seed(42)
X_train = np.linspace(-5, 5, 100).reshape(-1, 1)
y_train = 1 / (1 + np.exp(-X_train)) + np.random.normal(0, 0.1, size=X_train.shape)

# Build the neural network model
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,), dtype=tf.float32),
    tf.keras.layers.Dense(units=10, activation='sigmoid'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(X_train, y_train, epochs=500, verbose=0)

# Generate predictions
X_test = np.linspace(-7, 7, 100).reshape(-1, 1)
y_pred = model.predict(X_test)

# Plot the results
plt.plot(X_train, y_train, 'o', label='Training Data')
plt.plot(X_test, y_pred, label='Neural Network Prediction')
plt.title('Neural Network with Sigmoid Activation')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()

```
